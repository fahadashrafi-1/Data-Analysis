```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Some Random Data
# You can use Faker to generate some data


df = pd.DataFrame({'ProjectID' : ['Proejct-1', 'Proejct-2', 'Proejct-3', 'Proejct-4', 'Proejct-5'],
                     'Revenu' : [55667, 35150, 98484, 62888, 37789]}
)

df
```

```python
fig, ax = plt.subplots()

x = df['ProjectID']
y = df['Revenu']

ax.bar(x , y, color='lightgreen', edgecolor='darkgreen')
ax.spines['bottom'].set_color('purple')

# For Making the Brars in Ascending Order

df = df.sort_values(by=['Revenu'])


for bars in ax.containers:
    ax.bar_label(bars)

plt.ylabel('Revenu')

plt.title('Top Projects by Revenu')
```
