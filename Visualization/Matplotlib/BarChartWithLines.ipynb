```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Some Random Data
# You can use Faker to generate some data


df = pd.DataFrame({'ProjectID' : ['Proejct-1', 'Proejct-2', 'Proejct-3', 'Proejct-4', 'Proejct-5'],
                     'Revenue' : [55667, 35150, 98484, 62888, 37789]}
)

fig, ax = plt.subplots()

x = df['ProjectID']
y = df['Revenue']

ax.bar(x , y, color='#8ADF98', edgecolor='darkgreen')
ax.spines['bottom'].set_color('purple')

# For Making the Brars in Ascending Order

df = df.sort_values(by=['Revenue'])


for bars in ax.containers:
    ax.bar_label(bars)

plt.ylabel('Revenue')

plt.title('Top Projects by Revenue')
```
