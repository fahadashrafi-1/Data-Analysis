```python
dataset = pd.read_excel('procurmentdata.xlsx')
```

#### The code below are the simple visiuals with are the components of a dashboard used in the filenamed Dashboard.ipynb

```python
# Create a Card Visiual in Matplotlib. Its fun working 
fig, asx = plt.subplots(figsize=(2,1))

# Grabbing Colors for our condition if achievment is below 90 our value will be Red
colorB = '#f4a261'
colorA = '#ef767a'

# You can change the value to check the color
cutoffvalue = dataset['PoValue'].sum()/1000

# Our Conditions to cange the value
if cutoffvalue <= 90:
    colorDyanamic = colorB
else:
    colorDyanamic = colorA
    

# Code for writing visiual

asx.text(0.2,0.2,str(cutoffvalue), fontsize=38, color=colorDyanamic)
asx.set(title='Total \n Purchase Orders')
asx.axis('off')    

```

```python

# below variable names are self explonatiory
labels = dataset.value_counts(['PoType'], normalize=True).index.tolist()
sizes = dataset.value_counts(['PoType'], normalize=True).values.tolist()

explode = (0.1, 0)

# Create a figuer
fig, asx = plt.subplots(figsize=(2,1))

# Switch of the Axis
asx.axis('off')
asx.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=55, radius=2.5, colors=['#74c69d','#ffe8d6'] )

fig.show
```

```python
fig, asx = plt.subplots(figsize=(2,1))

# We want to calculate the KPI of Procurment set by Funtoso
# percentage of orders which were delivered on time

deliverrtime = dataset['DeliverStatus'].value_counts(normalize=True).loc['OnTime']

asx.text(0.2,0.2,str(deliverrtime*100)+' %', fontsize=28, color='#03a9fc')
asx.set(title='On Time Delivery')
asx.axis('off')   
```
