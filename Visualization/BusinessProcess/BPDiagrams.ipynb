### Keeping work clean , simple and understandable is very important.
### A very handy and easy tool which is on the top of graphiviz , Diagrams, is very helpful

```python
from diagrams import Diagram
from diagrams.programming.framework import Django, React
from diagrams.custom import Custom
from diagrams.azure.database import SQLDatabases
```

```python

graph_attr = {
    "fontsize": "12",
    "bgcolor": "oldlace"
 
}


with Diagram("Tech used in Mobile Application", 
             show=False,
             filename='ApplicationSetup',
             direction="TB",  
             graph_attr=graph_attr,
             ) as d:
    
    [   
        React("React (Frontend, Web / Andorid / IOS) ")  >> 
        Django("Mobile Application Backend (Integrated with D-ERP)") >> 
        SQLDatabases("Mirosoft SQL as Database") >>
        Custom("ERP", "customs/ERP.PNG")
        ]

d

```

```python
from graphviz import Digraph
# Create Digraph object

dot = Digraph()

graph_attr = {
    "fontsize": "12",
    "bgcolor": "oldlace"
 
}

dot = Digraph('Mobile Application Model', format='png', graph_attr=graph_attr)

dot.node('2','Mobile Frontend', shape='box', color='green')
dot.node('3','Mobile IOS Frontend', shape='box', color='green')
dot.node('4','Web Frontend', shape='box', color='green')
dot.node('5','Python Django Beckend', shape='cylinder', color='red')
dot.node('6','ERP System', shape='box', color='blue')
dot.node('A','Sales Purchase', shape='invhouse',  color='blue')
dot.node('B','CRM', shape='invhouse',  color='blue')

dot.node('8','Integrations if Required', shape='box',  color='red')
dot.node('1','Integrations (payment system, Ticketing system )', shape='egg', color='blue')

dot.edges(['25', '35', '45', '56',  '58', '6A', '6B', 'B1', 'A1'])

dot.render(filename='Application')

dot
```
