#### Find below a very small snip of code for creating process for Business Scorecards

```python
import pandas as pd

#  You can find details at https://graphviz.readthedocs.io/en/stable/manual.html
import graphviz

# For showing Images
from IPython.display import Image
```

```python
# You can use whatever format required , For My case Pdf is right choice.

dot = graphviz.Digraph(format='svg', graph_attr={'label': 'Balance Scorecard Financial Datasoruces'})


dot.node('A', 'Finance', shape='box', style="filled", fillcolor='#D69CB4')
dot.node('1', 'BS-F-001', shape='rarrow', style="filled", color='lightblue')
dot.node('2', 'BS-F-002', shape='rarrow', style="filled", color='lightblue')
dot.node('3', 'BS-F-003', shape='rarrow', style="filled", color='lightblue')
dot.node('5', 'DataSoruce-1', shape='rect', style="filled", fillcolor='#BFA29B')
dot.node('6', 'DataSoruce-2', shape='rect', style="filled", fillcolor='#D69CB4')
dot.node('7', 'DataSoruce-3', shape='box', style="filled", fillcolor='#D69CB4')
dot.node('8', 'DataSoruce-4', shape='box', style="filled", fillcolor='#D69CB4')
dot.node('b', 'Employee \n Table', shape='cylinder' ,  style="filled", fillcolor='#aff0cd')
dot.node('c', 'ERP \n Table', shape='cylinder' ,  style="filled", fillcolor='#aff0cd')
dot.node('e', 'Projects \n Table', shape='cylinder' ,  style="filled", fillcolor='#aff0cd')
dot.node('d', 'Power BI \n Dashboard ', shape='note' ,  style="filled", fillcolor='#ede3bb')


dot.edges(['A1','A2', 'A3', '15', '16', '25' , '26', '37', '38', '5b', '8c', '6c', 'bd',  'cd', '7e', 'ed'])
dot.attr(rankdir='LR')  
dot.render(filename='BalanceScoreCard-KPI')
dot
```
